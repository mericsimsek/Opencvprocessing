```python
#alfa*pixel1 + beta*pixel2 + Y  ağırlıklı bi toplam yapılır bi pixelin daha fazla temsil etmesini istiorsanız alfayı arttırırız
#harmanlamada alfa ve betaları güçlendirdikçe değişimler olur
imgs1=cv2.imread(r"C:\Users\meric\Desktop\Computer-Vision-with-Python\DATA\meric.jpg")
imgs2=cv2.imread(r"C:\Users\meric\Desktop\Computer-Vision-with-Python\DATA\watermark_no_copy.png")
```

```python
imgs1=cv2.cvtColor(imgs1,cv2.COLOR_BGR2RGB)
imgs2=cv2.cvtColor(imgs2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imgs1)
```

```python
imgs1.shape,imgs2.shape #birbirine yapıştırmak için aynı boyutta ve yapıda olmaları gerekir benzer yapılara getirecez1200 e 1200
```

```python
imgs1=cv2.resize(imgs1,((1200,1200)))
imgs2=cv2.resize(imgs2,((1200,1200)))
```

```python
#alfa*pixel1 + beta*pixel2 + Y 
blended=cv2.addWeighted(src1=imgs1,alpha=0.5,src2=imgs2,beta=0.5,gamma=0)
plt.imshow(blended)
```

```python
#alfa*pixel1 + beta*pixel2 + Y 
blended=cv2.addWeighted(src1=imgs1,alpha=0.5,src2=imgs2,beta=0.5,gamma=0)
plt.imshow(blended)
```

```python
blended=cv2.addWeighted(src1=imgs1,alpha=1,src2=imgs2,beta=0.2,gamma=1)
plt.imshow(blended)
```

```python
imgs2=cv2.resize(imgs2,((600,600)))
#şimdi boyutların farklı oldğu durum
largeimg=imgs1
smallimg=imgs2
x_offset=0 #başlangıç noktaları
y_offset=0
```

```python
largeimg.shape,smallimg.shape
```

```python
x_end=x_offset + smallimg.shape[1]#x
y_end=y_offset + smallimg.shape[0]#y
```

```python
 largeimg[y_offset:y_end,x_offset:x_end]=smallimg
```

```python
Bu kod, küçük bir görüntüyü (smallimg) büyük bir görüntünün (largeimg) üzerine yerleştirir. Şimdi bunu adım adım açıklayalım:

1. x_offset ve y_offset
Bu değişkenler, küçük görüntünün büyük görüntüde nereye yerleştirileceğini belirler.
x_offset = 0 ve y_offset = 0, küçük görüntünün büyük görüntünün sol üst köşesinden başlayarak yerleştirileceğini belirtir.
2. x_end ve y_end
x_end ve y_end değişkenleri, küçük görüntünün büyük görüntü üzerindeki son koordinatlarını hesaplar.
x_end = x_offset + imgs1.shape[1], küçük görüntünün genişliğini x_offset ile toplayarak küçük görüntünün genişlik sınırını bulur.
y_end = y_offset + imgs1.shape[0], küçük görüntünün yüksekliğini y_offset ile toplayarak küçük görüntünün yükseklik sınırını bulur.
3. largeimg[y_offset:y_end, x_offset:x_end] = smallimg
Bu satır, küçük görüntüyü büyük görüntünün belirtilen alanına yerleştirir.
Büyük görüntünün y_offset:y_end, x_offset:x_end bölgesi, küçük görüntü ile aynı boyutlara sahip olmalı, aksi halde hata alırsınız.
Sonuç olarak, küçük görüntü büyük görüntünün sol üst köşesine yerleştirilir.
Örnek:
smallimg boyutu: 100x100 piksel
largeimg boyutu: 500x500 piksel
Bu durumda, smallimg büyük görüntünün sol üst köşesindeki (0, 0) koordinatından başlayarak (100, 100) koordinatına kadar büyük görüntünün üstüne yerleştirilir.
```
