```python
#alfa*pixel1 + beta*pixel2 + Y  arlkl bi toplam yaplr bi pixelin daha fazla temsil etmesini istiorsanz alfay arttrrz
#harmanlamada alfa ve betalar glendirdike deiimler olur
imgs1=cv2.imread(r"C:\Users\meric\Desktop\Computer-Vision-with-Python\DATA\meric.jpg")
imgs2=cv2.imread(r"C:\Users\meric\Desktop\Computer-Vision-with-Python\DATA\watermark_no_copy.png")
```

```python
imgs1=cv2.cvtColor(imgs1,cv2.COLOR_BGR2RGB)
imgs2=cv2.cvtColor(imgs2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imgs1)
```

```python
imgs1.shape,imgs2.shape #birbirine yaptrmak iin ayn boyutta ve yapda olmalar gerekir benzer yaplara getirecez1200 e 1200
```

```python
imgs1=cv2.resize(imgs1,((1200,1200)))
imgs2=cv2.resize(imgs2,((1200,1200)))
```

```python
#alfa*pixel1 + beta*pixel2 + Y 
blended=cv2.addWeighted(src1=imgs1,alpha=0.5,src2=imgs2,beta=0.5,gamma=0)
plt.imshow(blended)
```

```python
#alfa*pixel1 + beta*pixel2 + Y 
blended=cv2.addWeighted(src1=imgs1,alpha=0.5,src2=imgs2,beta=0.5,gamma=0)
plt.imshow(blended)
```

```python
blended=cv2.addWeighted(src1=imgs1,alpha=1,src2=imgs2,beta=0.2,gamma=1)
plt.imshow(blended)
```

```python
imgs2=cv2.resize(imgs2,((600,600)))
#imdi boyutlarn farkl oldu durum
largeimg=imgs1
smallimg=imgs2
x_offset=0 #balang noktalar
y_offset=0
```

```python
largeimg.shape,smallimg.shape
```

```python
x_end=x_offset + smallimg.shape[1]#x
y_end=y_offset + smallimg.shape[0]#y
```

```python
 largeimg[y_offset:y_end,x_offset:x_end]=smallimg
```

```python
Bu kod, kk bir grnty (smallimg) byk bir grntnn (largeimg) zerine yerletirir. imdi bunu adm adm aklayalm:

1. x_offset ve y_offset
Bu deikenler, kk grntnn byk grntde nereye yerletirileceini belirler.
x_offset = 0 ve y_offset = 0, kk grntnn byk grntnn sol st kesinden balayarak yerletirileceini belirtir.
2. x_end ve y_end
x_end ve y_end deikenleri, kk grntnn byk grnt zerindeki son koordinatlarn hesaplar.
x_end = x_offset + imgs1.shape[1], kk grntnn geniliini x_offset ile toplayarak kk grntnn genilik snrn bulur.
y_end = y_offset + imgs1.shape[0], kk grntnn yksekliini y_offset ile toplayarak kk grntnn ykseklik snrn bulur.
3. largeimg[y_offset:y_end, x_offset:x_end] = smallimg
Bu satr, kk grnty byk grntnn belirtilen alanna yerletirir.
Byk grntnn y_offset:y_end, x_offset:x_end blgesi, kk grnt ile ayn boyutlara sahip olmal, aksi halde hata alrsnz.
Sonu olarak, kk grnt byk grntnn sol st kesine yerletirilir.
rnek:
smallimg boyutu: 100x100 piksel
largeimg boyutu: 500x500 piksel
Bu durumda, smallimg byk grntnn sol st kesindeki (0, 0) koordinatndan balayarak (100, 100) koordinatna kadar byk grntnn stne yerletirilir.
```
