```python
1. Backpropagation (Geri Yayılım)
Backpropagation, sinir ağlarında modelin öğrenmesi için kullanılan bir yöntemdir. Bu yöntem, ağın çıktısı ile doğru çıktı arasındaki farkı (hata fonksiyonunu) minimize etmek amacıyla, gradyanların hesaplanmasını ve ağın her bir katmanındaki ağırlıkların güncellenmesini sağlar. İşleyişi şu şekildedir:

İlk önce, ileri besleme yapılır ve modelin çıktısı hesaplanır.
Daha sonra, çıktı ve beklenen sonuç arasındaki fark kullanılarak bir hata fonksiyonu (loss function) hesaplanır.
Bu hata fonksiyonuna göre, her bir ağırlık için türev alınır (gradyan hesaplanır) ve bu gradyanlar geri yayılır.
Ağırlıklar, bu gradyanlara göre güncellenir.
Gradyan, ağdaki her bir ağırlığın, model hatasını ne kadar etkilediğini gösterir. Eğer bir ağırlığın gradyanı büyükse, o ağırlık model hatasına büyük bir etki yapıyordur ve daha fazla güncellenmesi gerekir.

2. Gradient Descent (Gradyan Azaltma)
Gradient Descent, sinir ağlarını veya diğer makine öğrenmesi modellerini optimize etmek için kullanılan bir yöntemdir. Amacı, modelin hata fonksiyonunu (loss function) minimize etmektir. Bu işlem, fonksiyonun gradyanı boyunca en dik azalma yönünde ilerleyerek gerçekleştirilir.

Gradient descent şu şekilde çalışır:

Modelin hata fonksiyonu hesaplanır (örneğin, Mean Squared Error veya Cross-Entropy Loss).
Hata fonksiyonuna göre gradyanlar hesaplanır. Bu gradyanlar, hata fonksiyonunun eğimini ve hangi yönde azaldığını gösterir.
Gradyanlara göre modelin ağırlıkları güncellenir:

∇J(θ) ise hata fonksiyonunun gradyanını gösterir. Ağırlıklar, gradyanın ters yönünde adım adım azaltılır.
Gradyanların doğru bir şekilde hesaplanması, modelin doğru yöne ve doğru hızla ilerleyerek en iyi ağırlıkları öğrenmesini sağlar.

Gradyanlar, kenar bulma ve görüntüdeki önemli yapıların tespiti gibi görevler için kullanılır. Görüntülerde kenarlar, genellikle parlaklık değerlerinde ani değişiklikler ile ortaya çıkar. Gradyanlar, bu değişimlerin hızlı olduğu noktaları tespit etmek için kullanılır. Bu da, nesne tespiti, segmentasyon, görüntüdeki önemli alanların vurgulanması gibi görüntü işleme görevlerinde önemli bir adımdır.

Gradyan Hesaplamanın Yöntemleri (OpenCV Fonksiyonları)
Sobel Gradyanları

OpenCV'de gradyan hesaplamanın en yaygın yollarından biri Sobel operatörü kullanmaktır.
Sobel operatörü, bir görüntünün x ve y eksenlerindeki gradyanlarını ayrı ayrı hesaplar. Bu sayede, yatay ve dikey kenarları tespit edebilirsiniz.

x ve y ye göre filtreleme yapılabilir
```

```python
def displayimg(img):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
img=cv2.imread(r"C:\Users\meric\Desktop\Computer-Vision-with-Python\DATA\sudoku.jpg",0)
```

```python
displayimg(img)
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)#derinlik girilecek ve istenen çıkış derinliği  ve x 1  y 0 
```

```python
displayimg(sobelx)
```

```python
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)#derinlik girilecek ve istenen çıkış derinliği  ve x 0  y 1
```

```python
displayimg(sobely)
```

```python
aplacian Gradyanları

Laplacian operatörü, bir görüntünün ikinci türevini alarak gradyanları hesaplar. Bu, kenarların daha güçlü bir şekilde tespit edilmesini sağlar.
Laplacian, hem x hem de y eksenlerinde türevleri aynı anda hesapladığı için yatay ve dikey kenarları birlikte bulur.
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
displayimg(laplacian)
```

```python
Harmanlanmış görüntü (blended image), iki veya daha fazla görüntünün belirli bir yöntemle birleştirilerek
tek bir görüntü haline getirilmesidir. Harmanlama işlemi, genellikle iki görüntü arasındaki piksellerin belirli oranlarda
karıştırılması ile yapılır ve bu süreçte görüntülerin belirli bölgeleri birbirine geçiş yapar.
Bu teknik, grafik tasarım, sinema, artırılmış gerçeklik (AR), bilgisayarla görme (computer vision) ve görüntü işleme gibi 
birçok alanda kullanılır.

Harmanlanmış Görüntünün Amacı ve Kullanımı
Geçiş efekti oluşturmak: Harmanlama, iki görüntü arasında yumuşak bir geçiş oluşturmak için kullanılır.
    Bu, özellikle video geçişlerinde veya iki sahne arasında kesintisiz bir geçiş efekti elde etmek için önemlidir.

Görüntü birleştirme: Farklı görüntüleri harmanlayarak daha estetik veya bilgi açısından daha zengin görüntüler elde edilebilir. 
    Örneğin, bir manzara fotoğrafı ile bir portre fotoğrafını harmanlayarak yaratıcı kompozisyonlar oluşturabilirsiniz.

Dikişleme (Image Stitching): Birden fazla fotoğrafın bir araya getirilerek geniş bir panorama görüntüsü oluşturulmasında 
    harmanlama teknikleri kullanılır. Bu işlem sırasında fotoğrafların geçiş bölgelerinde pikseller düzgünce harmanlanarak 
    tek bir birleşik görüntü elde edilir.

Artırılmış Gerçeklik: Gerçek dünya görüntüleri ile bilgisayar tarafından üretilen görüntülerin birleştirilmesi 
    de bir harmanlama işlemidir. Bu sayede gerçek görüntü üzerine sanal nesneler yerleştirilebilir.
```

```python
blended=cv2.addWeighted(src1=sobelx,alpha=0.7,src2=sobely,beta=0.3,gamma=0) #yatayı daha net gösterttim ama normalde
#birbirlerine harmanlayıp daha net pikseller elde etmeyi amaçlayabilir
displayimg(blended)
```

```python
ret,th=cv2.threshold(img,100,255,cv2.THRESH_BINARY)
displayimg(th) #kenarları algılamaya çalışır
```

```python
kernel=np.ones((4,4),np.uint8)
```

```python
gradient=cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
displayimg(gradient)  #morph_gradient  kenarları vurgulamak içi
```

```python
blended: Morfolojik gradient işlemi uygulanacak görüntü. Bu durumda harmanlanmış bir görüntü (blended image) kullanıyorsunuz.
    
cv2.MORPH_GRADIENT: Morfolojik gradient işlemini tanımlar. 
    Bu işlem, görüntünün dilate (genişletme) ve erode (aşındırma) işlemleri arasındaki farkı hesaplar.
    Sonuç olarak, görüntüdeki kenarlar vurgulanır.
    
kernel: Morfolojik işlemde kullanılacak yapılandırma elemanı (kerneldir).
    Bu, genellikle bir matris şeklinde olur ve görüntüde hangi bölgelere işlem yapılacağını belirler.
    
displayimg(gradient):
Bu fonksiyon, elde edilen gradient görüntüsünü görüntülemek için kullanılır. Yani morfolojik gradient uygulandıktan sonra sonucu ekranda görmenizi sağlar.
Neden Yaptınız?

Morfolojik gradient uygulamanızın amacı:

Kenarları vurgulamak: Gradient işlemi, görüntüdeki nesnelerin kenarlarını vurgular ve çıkarır.
    Dilate ve erode işlemleri arasında fark alındığı için kenarlar belirgin hale gelir.
Gürültüye dayanıklı kenar çıkarımı: Diğer kenar bulma yöntemlerine kıyasla, morfolojik gradient işlemi gürültüye
    karşı daha dayanıklı olabilir, çünkü kenar tespiti için doğrudan piksellerle değil, yapılandırma elemanı ile işlem yapılır.
Kontur çıkarma: Görüntüdeki nesnelerin sınırlarını ve konturlarını daha belirgin hale getirmek için bu yöntemi kullanmış olabili
    rsiniz.
Bu işlem, görüntüdeki nesneleri ve arka planı ayırmada veya belirli bölgelerdeki özellikleri (kenar, sınır) çıkarmada
sıkça kullanılır. Gradient sonucu, nesnelerin kenarlarına odaklanarak onları daha net hale getirmek için idealdir.
```
